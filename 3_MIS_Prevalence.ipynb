```python
%%R
# for all loaded survey data, prepare required data columns and save results as a prevR object
# for the calculation of prevelance maps with the help of R libarary prevR, we require household (case) number,
# cluster number and additionally the variable we want to produce prevalence maps from, e.g. malaria cases,
# indoor residual spraying or distribution of bednets
for(i in 1:length(dataList)){
        df <- data.frame(dataList[i])
        # get columns of malaria rapid test cases --> either columns 21 to 53 (2011) or 20 to 40 (2013)
        if(ncol(df)>40){
                malariaAgg <- df[,21:53]      
        } else{
                malariaAgg <- df[,20:40]  
        }
        
        # Just utilise valid positive malaria cases --> 6 indicates neither positive nor negative
        malariaAgg[malariaAgg==6] <- NA
        
        # Calculate row sums of malaria rapid test casess
        malAgg_Sum <- rowSums(malariaAgg,na.rm=TRUE)
        
        # Variable HV253 responds whether the dwelling has been sprayed against mosquitoes after 12 months
        IRS <- df$HV253
        # --> 8 indicates "Don't know" and is excluded from our analysis
        IRS[IRS==8] <- NA
        # create dataframe with necessary information
        # --> HHID - Case identification
        # --> HV001 - Cluster number
        # --> additional prevalence variables, here: malaria cases, mosquito nets, IRS
        df_malAgg <- data.frame('HHID'=df$HHID, 'HV'=df$HV001, 'malAgg_sum'=c(malAgg_Sum),
                                'bedNets'=df$HV227,'IRS'=c(IRS))
        
        # rename loaded shapefile based on year of MIS
        if(i==1){
                geoLoc <- geoLoc_1  
        } else{
                geoLoc <- geoLoc_2
        }
        # Aggregate malaria rapid test data per cluster number
        # For prevelance maps, we need: 
        # (i) n: number of valid observations (total number of surveyed people per cluster)
        # (ii) pos: number of positive cases, e.g. positive tested malaria cases
        
        # 1. aggregate data and get number of valid observations per cluster (n)
        # 2. aggregate data and get number of positive cases per cluster (pos)
        
        # --> create final data frame for malaria prevalence
        ##################################################################
        dfMal.final <- merge(aggregate(malAgg_sum ~ HV, df_malAgg,length),
                             aggregate(malAgg_sum ~ HV, df_malAgg, sum), by='HV')
        
        # add geo location information to every cluster number
        dfMal.final <- cbind(dfMal.final,geoLoc@coords)
        # change column names
        colnames(dfMal.final) <- c('ClusterNr', "validObs", 'posCases',"lon",'lat')
                
        # --> create final data frame for prevalence of IRS
        ##################################################################
        dfIRS.final <- merge(aggregate(IRS ~ HV, df_malAgg,length),
                             aggregate(IRS ~ HV, df_malAgg, sum), by='HV')
        # add geo location information to every cluster number
        dfIRS.final <- cbind(dfIRS.final,geoLoc@coords)
        # change column names
        colnames(dfIRS.final) <- c('ClusterNr', "validObs", 'posCases',"lon",'lat')
        
        # --> create final data frame for prevalence of mosquito nets
        #################################################################
        dfbedNets.final <- merge(aggregate(bedNets ~ HV, df_malAgg,length),
                                 aggregate(bedNets ~ HV, df_malAgg, sum), by='HV')
        # add geo location information to every cluster number
        dfbedNets.final <- cbind(dfbedNets.final,geoLoc@coords)
        # change column names
        colnames(dfbedNets.final) <- c('ClusterNr', "validObs", 'posCases',"lon",'lat')
        
        
        # Identify valid columns for input for creating a prevR object
        # Needed is:
        # id - cluster number, x - longitude, y - latitude, n - number of valid observations
        # pos - number of positive cases
        col <- c(id='ClusterNr', x='lon', y='lat', n='validObs', pos='posCases')
        
        # create prevR objects for three prevalence variables
        prevR_mal_temp <- as.prevR(data=dfMal.final,col=col,boundary=MDG_boundary)
        prevR_IRS_temp <- as.prevR(data=dfIRS.final,col=col,boundary=MDG_boundary)
        prevR_bedNets_temp <- as.prevR(data=dfbedNets.final,col=col,boundary=MDG_boundary)
        
        # set rings for kernel estimation --> based on empirical equation after
        # Lamarange et al. 2011
        # Formula: No = 14.172*n^0.419*p^(-0.361)*g^0.037-91.011
        # n = number of persons tested
        # p = observed national prevalence
        # ,g = number of clusters of the survey
        prevR_mal_temp <- rings(prevR_mal_temp,N=Noptim(prevR_mal_temp))
        prevR_IRS_temp <- rings(prevR_IRS_temp,N=Noptim(prevR_IRS_temp))
        prevR_bedNets_temp <- rings(prevR_bedNets_temp,N=Noptim(prevR_bedNets_temp))
        
        # rename prevR objects related to MIS survey
        assign(paste('prevR_mal_',i,sep=''),prevR_mal_temp)
        assign(paste('prevR_IRS_',i,sep=''),prevR_IRS_temp)
        assign(paste('prevR_bedNets_',i,sep=''),prevR_bedNets_temp)
}
```

```python
%%R
# Put individual prevR objects to lists
prevList_1 <- list(prevR_mal_1,prevR_IRS_1,prevR_bedNets_1)
prevList_2 <- list(prevR_mal_2,prevR_IRS_2,prevR_bedNets_2)

# final list that contains all prevR objects to interpolate
prevList <- list(prevList_1,prevList_2)

prevMappingToNCDF <- function(prevList,varName,fileName){
###############################################################################
# Interpolates prevR objects (given by the input list) with a kernel estimator approach
# (after Larmarange et al. 2011), puts the interpolated surfaces into a raster stack and 
# creates a netCDF out of the raster stack.
#
# Args:  
#   prevList:   list that contains various lists of prevR objects that shall be 
#               interpolated
#   varName:    name of netCDF variable
#   fileName:   fileName specifies the name of the resulting netCDF file
	
#
# Returns:
#	nothing is returned --> output is written to a netCDF file
###############################################################################
        for(j in 1:length(prevList)){
                tempList <- prevList[[j]]
                for(i in 1:length(tempList)){
                        test.kde <- kde(tempList[[i]],N=Noptim(tempList[[i]]),weighted=FALSE,risk.ratio=FALSE,
                                        keep.details=FALSE, cell.size=0.01,progression=TRUE)
                        lat <- sort(unique(test.kde$y),decreasing=TRUE)
                        lon<- unique(test.kde$x)
                        if(i==1){
                                s<- stack(raster(test.kde))
                        } else{
                                s<- addLayer(s,raster(test.kde))
                        }        
                }
                names(s) <- c('prevMalaria', 'prevIRS', 'prevBedNets')
                writeStackToNCDF(stack = s,lat=lat,lon=lon,timeUnit='prevLayer',
                                 varName=varName,
                                 varUnit='%',
                                 varDescription='This stack consists of three prevalence maps: malaria, IRS and bedNets. Prevalence maps have been calculated with the help of the R library prevR and data from Malaria Indicator Survey from USAID DHS.',
                                 timeStepVec=c(1:3),fileName=fileName)
        }
}

# Call the function
prevMappingToNCDF(prevList=prevList,varName='Prevalence maps for madagascar 2013',fileName='MIS_2013_prepared')
```
