## Downscaling guide for TRMM Precipitation Data

This notebook has been written to help downscaling TRMM precipitation data  from a 0.25° resolution to 1 km resolution, with the help of open source applications.

### Requirements

Input Files:

- NDVI image (1km resolution or better)
- DEM image (1km resolution or better)
- TRMM Precipitation image (0,25° resolution)


Software:

- **QGIS**, freely available [here](https://www.qgis.org/en/site/forusers/download.html)
- **GWR4**, for geographically weighted regression, freely available [here](https://geodacenter.asu.edu/software/downloads/gwr_downloads)

### Overview

The procedure presented in this notebook follows the article of Chen et al. (2014): spatial downscaling of TRMM 3B43 precipitation, available [here](http://www.tandfonline.com/doi/abs/10.1080/01431161.2014.902550#.VWMHDPmqpBc).

The steps listed in the article are:

1. Resample NDVI annual average values ($NDVI_{1km}$) and $DEM_{1km}$ from 1 km to 0.25° by pixel averaging.
2. Use GWR to establish an empirical functional relationship among $TRMM$, $DEM_{0.25°}$, and $NDVI_{0.25°}$ for a certain year and a particular study area, which is expressed in the function:

    $TRMM = \beta_0(u) + \beta_1(u) \times DEM_{0.25°} + \beta_2(u) \times NDVI_{0.25°} + \epsilon(u)$

3. Resample the model parameters ($\beta_0(u), \beta_1(u)$ and $\beta_2(u)$) to 1 km by the nearest resampling method
4. Estimate high-resolution annual precipitation (1 km) from $NDVI_{1km}$ and $DEM_{1km}$ using the model parameters derived in step (3.), which is expressed as:

    $p_{estimated}^{1km}(u)=\beta_0^{1km}(u)+\beta_1^{1km}(u)\times DEM_{1km} +\beta_2^{1km}(u)\times NDVI_{1km}$

5. Interpolate the residual $\epsilon_{0.25°}(u)$ into a grid of 1 km pixels using interpolating methods such as inverse distance weighting (Shepard 1968), ordinary kriging (Goovaerts 1997), and spline interpolation (Franke 1982).

6. The final downscaled precipitation is generated by adding the estimated values of $p_{estimated}^{1km}(u)$ to the residual correction term $\epsilon_{1km}(u)$ obtained in step (5.), and is written as 

    $p_{downscaled}^{1km}(u)=p_{estimated}^{1km}(u)+\epsilon_{1km}(u)$

### Step 1. Resampling

To resample NDVI and DEM images we can use **GDAL_TRANSLATE**, it can be used:

- From **command line**("gdal_translate -parameters").
- From **QGIS** under Raster -> Conversion -> Translate.
- From the iPython notebook **Resampling.ipynb**, which is set up to resample images from any resolution to 0.25°.

In any case, the only parameter to work with is **outsize**, it represents the size of the output image.

The formula to obtain the correct value for **outsize** is:

 $outsize=\frac{Pixelsize_{NDVIorDEM}\times 100}{0.25}$
---



However, if the result is a decimal number, there may be problems depending on which way **GDAL_TRANSLATE** is called:

- **QGIS** doesn't accept decimal numbers at all
- **command line** doesn't accept decimal numbers between 0 and 1
- **Resampling.ipynb** is already set to overcome this problem

There is a workaround to this problem, that is calling **GDAL_TRANSLATE** multiple times:

- **outsize** has to be multiplied by 10 for every decimal digit we want to consider in the approximation
- **GDAL_TRANSLATE** has to be called multiple times with **outsize** 10, the number of times is equal to the number of multiplications by 10 that has been done in the step before.
- **GDAL_TRANSLATE** has to be called one last time with the integer part of the **outsize** that has been calculated in the first step of the workaround

Example: if the result of **outsize** is 3.6, the algorithm will be called once with **outsize**=10 and another time with **outsize**=36

### Step 2. Geographically Weighted Regression

#### QGIS

- Open **QGIS** and add the NDVI and DEM resampled images along with the TRMM precipitation image
- Choose *Vector -> Research Tools -> Regular Points* , choose the TRMM file as *Input Boundary Layer* and select *Use this number of points* with 2000 as value.
- Install the plugin *Point Sampling Tool* from the menu *Plugins -> Manage and Install plugins...*
- Choose *Plugins -> Analyses -> Point Sampling Tool*, select the layer containing 2000 points and the 3 rasters (NDVI,DEM,TRMM)

Now we have a vector containing 2000 points, each point contains information about the 3 rasters and the coordinates(x,y).

However, we have to take a subset of valid points, since most contain NaN values:
- Right click on the last added vector layer and choose *Filter...*
- In the Query Builder, we should have 3 Fields, each with its own *fieldname* e.g. *ndvi*,*dem*,*trmm*.
- Insert this expression with the correct field names: 

    "*fieldname1*" IS NOT NULL AND "*fieldname2*" IS NOT NULL AND "*fieldname3*" IS NOT NULL

- Click on *Test* to see if it returns any row and then click on ok
- Now Right click on the vector layer and choose *Save as...* with format type CSV
- Open the .csv file with a text editor(e.g. Notepad++) and save it as a .txt file

#### GWR4

- Open **GWR4** and load the .txt file
- In *Step 2* assign the **X** and **Y** variables from the *Variable list* to the *Locational variables*
- Assign the **DEM** and **NDVI** variables from the *Variable list* to the *Local Independent variables*
- Assign the **TRMM** variable from the *Variable list* to the *Dependent variable*
- Configure *Step 3*
- In *Step 4* choose the name for the output files
- In *Step 5* execute the **GWR**

#### QGIS

Now that we have $\beta_0(u), \beta_1(u), \beta_2(u), \epsilon(u)$ we can continue with the procedure on **QGIS**:

- From the column on the left, select *Add delimited text layer* and choose the file *_listwise.csv*.
- From the menu *Raster -> Interpolation -> Interpolation* select the input layer *_listwise* and choose *residual* in Interpolation Attribute, insert in *Cellsize X* and *Cellsize Y* the value 0.25.
- Repeat the previous step for each Interpolation Attribute that has *"est_"* in the name.

### Step 3. Resample $\beta_0(u), \beta_1(u), \beta_2(u), \epsilon(u)$ to 1km

To complete this step, we interpolate the *listwise* vector with the value 0.01, so repeat the interpolation process explained above with *Cellsize X* and *Cellsize Y* = 0.01

### Step 4. Estimate $p_{estimated}^{1km}(u)$

In **QGIS** from the menu *Raster* choose *Raster Calculator* and insert the formula: 

$p_{estimated}^{1km}(u)=\beta_0^{1km}(u)+\beta_1^{1km}(u)\times DEM_{1km} +\beta_2^{1km}(u)\times NDVI_{1km}$

Where:
- $\beta_0^{1km}(u)$ is the estimated intercept raster
- $\beta_1^{1km}(u)$ is the estimated DEM raster
- $\beta_2^{1km}(u)$ is the estimated NDVI raster
- $DEM_{1km}$ is the original 1km DEM raster
- $NDVI_{1km}$ is the original 1km NDVI raster

### Step 6. Calculate $p_{downscaled}^{1km}(u)$

- $p_{estimated}^{1km}(u)$ is the raster calculated during Step 4.
