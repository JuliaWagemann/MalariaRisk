### Monthly extraction from ncdf

```python
%%px
%%R -i month,monthName
library(raster)
library(rgdal)
path_evap <- "P:/3_Malaria_Risk/1_Data/9_Evapotranspiration"

setwd (path_evap)
myshp <- readOGR("C:/Alex/script/MDG_adm0.shp", layer=basename("MDG_adm0") )
e <- extent(myshp)
fileList <- list.files(path=path_evap, full.names = FALSE, pattern = "*.nc")

for (i in fileList){
myraster <- raster(i, band = month+1)

# Cropping the raster to the shapefile spatial extent
myraster.crop <- crop(myraster, e, snap="out")

# Dummy raster with a spatial extension equal to the cropped raster,
# but full of NA values
crop          <- setValues(myraster.crop, NA)

#  Rasterize the catchment boundaries, with NA outside the catchment 
myshp.r <- rasterize(myshp, crop)

# Putting NA values in all the raster cells outside the shapefile boundaries
myraster.masked <- mask(x=myraster.crop, mask=myshp.r)

#plot(myraster)
#print(myraster.masked)
outputname <- i

outputname <- sub(".nc" , monthName, outputname)
outfile <- paste("C:/Alex/script/evapo/",monthName, sep="")
outfile2 <- paste(outfile, "/", sep="")
outfile3 <- paste(outfile2, outputname, sep="")
outfile4 <- paste(outfile3, ".tif", sep="")
writeRaster(myraster.masked, outfile4, format = "GTiff")
#plot()
}
```

### Load already processed 2011 and 2013 images and process them via Processing-GRASS r:series

```python

tiflist2011= '' 
tiflist2013= ''
for path, subdirs, files in os.walk(wd):
    for name in files:
        filename=os.path.join(path, name)
        #print (filename)
        if(filename.count("2011")&filename.endswith(".tif")):
            tiflist2011=tiflist2011+filename+";"
        if(filename.count("2013")&filename.endswith(".tif")):
            tiflist2013=tiflist2013+filename+";"
tiflist2011=tiflist2011[:-1]
tiflist2013=tiflist2013[:-1]
print(tiflist2011)
print(tiflist2013)
templayer=QgsRasterLayer(tiflist2011.split(";")[0])
temp=templayer.extent()
extent=str(temp.xMinimum())+','+str(temp.xMaximum())+','+str(temp.yMinimum())+','+str(temp.yMaximum())

processing.runalg('grass:r.series',tiflist2011,True, 4, '-10000000000,10000000000',extent,0, wd+'2011/'+country+'_Evapo_2011_Minimum.tif')
processing.runalg('grass:r.series',tiflist2011,True, 6, '-10000000000,10000000000',extent,0, wd+'2011/'+country+'_Evapo_2011_Maximum.tif')
processing.runalg('grass:r.series',tiflist2011,True, 0, '-10000000000,10000000000',extent,0, wd+'2011/'+country+'_Evapo_2011_Average.tif')
#processing.runalg('grass:r.series',tiflist2011,True, 10, '-10000000000,10000000000',extent,0, wd+'2011/'+country+'_Evapo_2011_Sum.tif')

processing.runalg('grass:r.series',tiflist2013,True, 4, '-10000000000,10000000000',extent,0, wd+'2013/'+country+'_Evapo_2013_Minimum.tif')
processing.runalg('grass:r.series',tiflist2013,True, 6, '-10000000000,10000000000',extent,0, wd+'2013/'+country+'_Evapo_2013_Maximum.tif')
processing.runalg('grass:r.series',tiflist2013,True, 0, '-10000000000,10000000000',extent,0, wd+'2013/'+country+'_Evapo_2013_Average.tif')
#processing.runalg('grass:r.series',tiflist2013,True, 10, '-10000000000,10000000000',extent,0, wd+'2013/'+country+'_Evapo_2013_Sum.tif')
```
